```python
import pandas as pd
df = pd.read_csv('EURUSD60.csv')
```

```python
df.shape
```

```python
df.head()
```

```python
df.rename( columns={ '1971.01.04':'date', '00:00':'time' , '0.53690':'open', '0.53690.1':'high', '0.53690.2':'low', '0.53690.3':'close', '1':'volume' }, inplace=True )
```

```python
df
```

```python
df['datetime'] = df['date'] + ' ' + df['time']
```

```python
df['datetime'] = pd.to_datetime( df['datetime'], format='%Y.%m.%d %H:%M' )
```

```python
df
```

```python
df = df[ ['datetime', 'time', 'open', 'high', 'low', 'close'] ]
```

```python
df.head()
```

```python
filt = ( df['datetime'] >= pd.to_datetime('2013-01-01') )
df_new = df.loc[filt]
```

```python
df_new = df_new[['datetime', 'open', 'high', 'low', 'close']]
```

```python
df_new.reset_index(inplace=True)
```

```python
df_new = df_new[['datetime', 'open', 'high', 'low', 'close']]
```

```python
df_new.rename( columns={ 'datetime': 'datetime', 'open':'open_1', 'high':'high_1', 'low':'low_1', 'close':'close_1' }, inplace=True )
```

```python
df_new
```

```python
df_new['close_2'] = df_new['close_1'].shift(1)
```

```python
df_new['close_3'] = df_new['close_2'].shift(1)
```

```python
df_new['EMA50'] = df_new.iloc[:,4].ewm(span=50,adjust=False).mean()
```

```python
df_new['open_2'] = df_new['open_1'].shift(1)
```

```python
df_new['open_3'] = df_new['open_2'].shift(1)
```

```python
df_new['high_2'] = df_new['high_1'].shift(1)
```

```python
df_new['high_3'] = df_new['high_2'].shift(1)
```

```python
df_new['low_2'] = df_new['low_1'].shift(1)
```

```python
df_new['low_3'] = df_new['low_2'].shift(1)
```

```python
df_new = df_new[ [ 'datetime', 'open_1', 'open_2', 'open_3', 'high_1', 'high_2', 'high_3', 'low_1', 'low_2', 'low_3', 'close_1', 'close_2', 'close_3' ]]
```

```python
df_new
```

```python
df_new['EMA50_1'] = df_new.iloc[:,10].ewm(span=50,adjust=False).mean()
```

```python
df_new['EMA50_2'] = df_new['EMA50_1'].shift(1)
```

```python
df_new['EMA50_3'] = df_new['EMA50_2'].shift(1)
```

```python
df_new['EMA21'] = df_new.iloc[:,10].ewm(span=21,adjust=False).mean()
```

```python
df_new['EMA21_2'] = df_new['EMA21'].shift(1)
```

```python
df_new['EMA21_3'] = df_new['EMA21_2'].shift(1)
```

```python
df_new.rename( columns={'EMA21':'EMA21_1'} , inplace=True )
```

```python
df_new.info()
```

```python
df_new.dropna(axis='index', how='any', inplace=True)
```

```python
df_new
```

```python
df_new['date'] = pd.to_datetime(df_new['datetime']).dt.date
```

```python
df_new['hour'] = pd.to_datetime(df_new['datetime']).dt.hour
```

```python
df_new['day'] = pd.to_datetime(df_new['datetime']).dt.dayofweek
```

```python
df_new.set_index('datetime', inplace=True)
```

```python
df_new.to_csv('eurusd_h1_2013_2020.csv')
```
